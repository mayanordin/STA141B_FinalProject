```python


# Loop through pages 1 to 361
for page in range(1, 361):  # Loop from 1 to 361
    # Parse the page source with BeautifulSoup
    soup = BeautifulSoup(driver.page_source, 'html.parser')

    # Extract all job post divs
    job_divs = soup.find_all('div', class_='job-card-list__entity-lockup')

    # Loop through each job post div
    for div in job_divs:
        # Extract the job title relative to the current div
        title_element = div.find('a', class_='ember-view')
        job_title = title_element.text.strip() if title_element else 'No title available'

        # Extract the company name relative to the current div
        company_element = div.find('span', class_='cSgAFzYHhlVAudlPUBRKZppIxOTCPQFaRzd')
        company_name = company_element.text.strip() if company_element else 'No company available'

        # Extract the location relative to the current div
        location_element = div.find('li', class_='job-card-container__metadata-item')
        location = location_element.text.strip() if location_element else 'No location available'

        # Extract the salary relative to the current div
        salary_element = div.find_all('span', dir='ltr')  # Look for all <span dir="ltr">
        salary = 'No salary available'
        for span in salary_element:
            if '$' in span.text:  # Check if the text contains a salary indicator
                salary = span.text.strip()
                break

        # Determine the mode of work (remote, hybrid, etc.)
        mode_of_work = "On-site"  # Default assumption
        if 'Remote' in location:
            mode_of_work = 'Remote'
        elif 'Hybrid' in location:
            mode_of_work = 'Hybrid'

        # Append the extracted data to the posts list
        posts.append({
            'title': job_title,
            'company': company_name,
            'location': location,
            'salary': salary,
            'mode_of_work': mode_of_work
        })
    
    # After scraping the current page, click the "Next Page" button
    try:
        # Find the "Next Page" button using XPath
        next_button = driver.find_element(By.XPATH, f'//button[@aria-label="Page {page + 1}"]')
        next_button.click()  # Click the "Next Page" button
        time.sleep(20)  # Wait for the next page to load (adjust sleep time as necessary)
    except:
        print(f"Error: Unable to find the next page button on page {page}.")
        break  # Stop if no "Next Page" button is found or if we've reached the last page

# Print the results
for post in posts:
    print(post)


```

```python
# Function to remove duplicate job titles
def clean_title(title):
    # Find the length of the title
    title_length = len(title)
    
    # Try to find the repeating part of the title and remove it
    for i in range(1, title_length // 2 + 1):
        if title[:i] == title[i:i + i]:
            return title[:i]  # Return the first part without the repetition
    return title  # If no duplication is found, return the title as is

# Apply the function to the 'title' column
df10['title'] = df10['title'].apply(clean_title)
```

```python
import re

# Function to remove words like "On-site", "Hybrid", "Remote" and parentheses from location
def clean_location(location):
    # Use regex to remove 'On-site', 'Hybrid', or 'Remote' and anything in parentheses
    cleaned_location = re.sub(r'\s*(On-site|Hybrid|Remote|[\(\)]).*\s*', '', location, flags=re.IGNORECASE)
    return cleaned_location.strip()  # Remove any leading/trailing spaces

# Apply the function to the 'location' column
df10['location'] = df10['location'].apply(clean_location)

```
